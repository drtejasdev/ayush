```python
import pandas as pd
```

```python
df = pd.read_excel('/content/iihmr dataset.xlsx')
```

```python
df.head(1000)
```

```python
total_patients = df['Patient ID'].nunique()
print("total sonu tapase patients:", total_patients)
```

```python
average_los = df['Length of Stay in days'].mean()
print(f"Average Length of stay: {average_los} days")
```

```python
unique_admission_types = df['Admission Type'].unique()
print("Unique Admission Types:", unique_admission_types)
```

```python
primary_diagnosis_distribution = df['Primary Diagnosis'].value_counts()
print("Primary Diagnosis Distribution:", primary_diagnosis_distribution)
```

```python
avg_age = df.groupby('Admission Type')['Age'].mean()
print(avg_age)
```

```python
critical_patients_long_stay = df[(df['Severity of Illness'] == 'Critical') & (df['Length of Stay in days'] > 30)]
print("Number of critical patients with LOS > 30 days:", len(critical_patients_long_stay))
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Length of Stay in days'], bins=20, edgecolor='white')
plt.title('Distribution of Length of Stay')
plt.xlabel('Length of Stay (days)')
plt.ylabel('Frequency')
plt.show()
```

```python
admission_type_counts = df['Admission Type'].value_counts()

admission_type_counts.plot(kind='pie', autopct='%1.4f%%', startangle=45, figsize=(10, 12), legend= True)
plt.title('Distribution of Patients by Admission Type')
plt.ylabel('')  # Remove the y-label for cleaner visualization
plt.show()
```

```python
admission_type_counts = df['Admission Type'].value_counts()

admission_type_counts.plot(kind='bar', figsize=(8, 8), edgecolor='black')
plt.title('Number of Patients by ARUN KUMAR Admission Type')
plt.xlabel('Admission Type')
plt.ylabel('Number of Patients')
plt.xticks(rotation=360)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Age'], df['Length of Stay in days'], alpha=)
plt.title('Age vs Length of Stay')
plt.xlabel('Age')
plt.ylabel('Length of Stay (days)')
plt.show()
```
